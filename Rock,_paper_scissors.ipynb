```python
import random
import ipywidgets as widgets
from IPython.display import display, clear_output
import time

# Initialize scores
player_score = 0
bot_score = 0
winning_score = 3

# Labels to display the scores and result
score_label = widgets.Label(value="Score - Player: 0 Bot: 0")
result_label = widgets.Label(value="")

# Function to animate the choices
def animate_choices(player_choice, bot_choice):
    animation_frames = [
        ("Rock", "Rock"),
        ("Paper", "Paper"),
        ("Scissors", "Scissors"),
        ("Rock", "Paper"),
        ("Paper", "Scissors"),
        ("Scissors", "Rock"),
        ("Rock", "Scissors"),
        ("Paper", "Rock"),
        ("Scissors", "Paper"),
    ]
    for frame in animation_frames:
        clear_output(wait=True)
        display(widgets.VBox([
            widgets.HBox([widgets.Label(value=f"Player: {frame[0]}"), widgets.Label(value=f"Bot: {frame[1]}")]),
            result_label,
            score_label,
            widgets.HBox([rock_button, paper_button, scissors_button])
        ]))
        time.sleep(0.1)
    clear_output(wait=True)
    display(widgets.VBox([
        widgets.HBox([widgets.Label(value=f"Player: {player_choice}"), widgets.Label(value=f"Bot: {bot_choice}")]),
        result_label,
        score_label,
        widgets.HBox([rock_button, paper_button, scissors_button])
    ]))

# Function to handle the player's choice
def play(player_choice):
    global player_score, bot_score
    bot_choice = random.choice(["Rock", "Paper", "Scissors"])
    animate_choices(player_choice, bot_choice)

    if player_choice == bot_choice:
        result = "Tie!"
    elif player_choice == "Rock":
        if bot_choice == "Paper":
            result = f"You lose! {bot_choice} covers {player_choice}"
            bot_score += 1
        else:
            result = f"You win! {player_choice} smashes {bot_choice}"
            player_score += 1
    elif player_choice == "Paper":
        if bot_choice == "Scissors":
            result = f"You lose! {bot_choice} cut {player_choice}"
            bot_score += 1
        else:
            result = f"You win! {player_choice} covers {bot_choice}"
            player_score += 1
    elif player_choice == "Scissors":
        if bot_choice == "Rock":
            result = f"You lose! {bot_choice} smashes {player_choice}"
            bot_score += 1
        else:
            result = f"You win! {player_choice} cut {bot_choice}"
            player_score += 1

    # Update the result and scores
    result_label.value = result
    score_label.value = f"Score - Player: {player_score} Bot: {bot_score}"

    # Check for a winner
    if player_score >= winning_score:
        result_label.value += "\nCongratulations! You win the game!"
        reset_game()
    elif bot_score >= winning_score:
        result_label.value += "\nSorry, the bot wins the game!"
        reset_game()

# Function to reset the game
def reset_game():
    global player_score, bot_score
    player_score = 0
    bot_score = 0
    score_label.value = "Score - Player: 0 Bot: 0"

# Create buttons for player choices
rock_button = widgets.Button(description="Rock")
paper_button = widgets.Button(description="Paper")
scissors_button = widgets.Button(description="Scissors")

# Link buttons to the play function
rock_button.on_click(lambda x: play("Rock"))
paper_button.on_click(lambda x: play("Paper"))
scissors_button.on_click(lambda x: play("Scissors"))

# Display buttons and initial scores
display(widgets.VBox([
    score_label,
    result_label,
    widgets.HBox([rock_button, paper_button, scissors_button])
]))
```
